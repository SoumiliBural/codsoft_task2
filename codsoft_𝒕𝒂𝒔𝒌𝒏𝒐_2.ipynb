```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from CSV
# Adjust sep to ',' assuming columns are separated by commas
df_movie = pd.read_csv('/IMDB MOVIES.csv', sep=',', engine='python', encoding='latin-1')

# Print the column names to verify
print(df_movie.columns)

# Plotting a Bar Chart
plt.figure(figsize=(12, 8))
sns.countplot(data=df_movie, x='Genre', palette='viridis')
plt.title('Number of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data preparation
df_final = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female'],
    'Ratings': ['Good', 'Bad', 'Good', 'Excellent', 'Bad', 'Good']
})

# Calculate the count of each rating for each gender
rating_counts = df_final.groupby('Gender')['Ratings'].value_counts()

# Extracting counts for Male and Female ratings
male_counts = rating_counts['Male']
female_counts = rating_counts['Female']

# Plotting a Pie Chart for Male Ratings
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(male_counts, labels=male_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Male Ratings Distribution')

# Plotting a Pie Chart for Female Ratings
plt.subplot(1, 2, 2)
plt.pie(female_counts, labels=female_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Female Ratings Distribution')

plt.tight_layout()
plt.show()
```
